```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data.csv', encoding='utf-8')
data
```

```python
data_plot = data.copy()
plt.scatter(data_plot['longitude'], data_plot['dimension'])
```

```python
data_TS = pd.read_csv('./TS.csv', encoding='utf-8')
data_DP = pd.read_csv('./DP.csv', encoding='utf-8')
data_Greedy = pd.read_csv('./Greedy.csv', encoding='utf-8')
data_RA = pd.read_csv('./RA.csv', encoding='utf-8')
data_TS, data_DP, data_Greedy, data_RA
```

```python
res_TS = data_TS.iloc[30:]['0'].tolist()
res_DP = data_DP.iloc[30:]['0'].tolist()
res_Greedy = data_Greedy.iloc[30:]['0'].tolist()
res_RA = data_RA.iloc[30:]['0'].tolist()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 9))
axs[0][0].scatter(data['longitude'], data['dimension'], color='#002C2F')
axs[0][0].plot(data_TS['longitude'], data_TS['dimension'], '#1D6A96')
axs[0][0].set_xlim(-100, 2050)
axs[0][0].set_ylim(300, 2450)
axs[0][0].set_title('TS')

axs[0][1].scatter(data['longitude'], data['dimension'], color='#002C2F')
axs[0][1].plot(data_DP['longitude'], data_DP['dimension'], '#1D6A96')
axs[0][1].set_xlim(-100, 2050)
axs[0][1].set_ylim(300, 2450)
axs[0][1].set_title('DP')

axs[1][0].scatter(data['longitude'], data['dimension'], color='#002C2F')
axs[1][0].plot(data_Greedy['longitude'], data_Greedy['dimension'], '#1D6A96')
axs[1][0].set_xlim(-100, 2050)
axs[1][0].set_ylim(300, 2450)
axs[1][0].set_title('Greedy')

axs[1][1].scatter(data['longitude'], data['dimension'], color='#002C2F')
axs[1][1].plot(data_RA['longitude'], data_RA['dimension'], '#1D6A96')
axs[1][1].set_xlim(-100, 2050)
axs[1][1].set_ylim(300, 2450)
axs[1][1].set_title('RA')

fig.tight_layout()
plt.savefig('./total-result.png', dpi=600)
plt.show()
```
